```
from vega_datasets import data
unemp_data = data.unemployment_across_industries()
unemp_data.head(100)
```

```
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
sns.lineplot(x='year', y ='rate', hue='series', alpha=0.5, data=unemp_data, hue_order=['Agriculture', 'Construction', 'Finance'])
import matplotlib.pyplot as plt
plt.title('unemployment rates over years \n')
plt.show()

```
