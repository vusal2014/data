```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 10))
lending_club_data = pd.read_csv('https://raw.githubusercontent.com/cjflanagan/sample_data/master/loan.csv')
lending_club_data.head()
sns.violinplot(x='loan_amnt', y='purpose', hue='term', data=lending_club_data, split=True)
plt.title('Purpose vs Loan Amount \n')
plt.legend()
```
